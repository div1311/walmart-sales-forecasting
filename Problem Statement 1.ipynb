# Check for stationarity by Dickey-Fuller test

```python
from statsmodels.tsa.stattools import adfuller

# Define a function to perform the Dickey-Fuller test and display the results
def test_stationarity(timeseries):
    result = adfuller(timeseries, autolag='AIC')
    print("Dickey-Fuller Test Results:")
    print(f'Test Statistic: {result[0]}')
    print(f'p-value: {result[1]}')
    print(f'Number of Lags Used: {result[2]}')
    print(f'Number of Observations Used: {result[3]}')
    print(f'Critical Values:')
    for key, value in result[4].items():
        print(f'   {key}: {value}')
    
    # Determine if the series is stationary based on the p-value
    if result[1] <= 0.05:
        print("\nResult: The time series is stationary.")
    else:
        print("\nResult: The time series is not stationary.")

test_stationarity(df10['Weekly_Sales'])
```

## Checking stationarity

```python
# Define a function to perform the Dickey-Fuller test and display the results
def test_stationarity(timeseries):
    result = adfuller(timeseries, autolag='AIC')
    print("Dickey-Fuller Test Results:")
    print(f'Test Statistic: {result[0]}')
    print(f'p-value: {result[1]}')
    print(f'Number of Lags Used: {result[2]}')
    print(f'Number of Observations Used: {result[3]}')
    print(f'Critical Values:')
    for key, value in result[4].items():
        print(f'   {key}: {value}')
    
    # Determine if the series is stationary based on the p-value
    if result[1] <= 0.05:
        print("\nResult: The time series is stationary.")
    else:
        print("\nResult: The time series is not stationary.")

test_stationarity(df14['Weekly_Sales'])
```
